```
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
```

```
x,y=make_blobs(
    n_samples=100,
    centers=8,
    cluster_std=0.8,
    random_state=40
    )
```

```
import numpy as np
db_scan=DBSCAN(eps=1.0, min_samples=5)
y_db=db_scan.fit_predict(x)
print(y_db)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], cmap='plasma', c=y_db, s=50)
plt.title('DBSCAN')
plt.show()
```
